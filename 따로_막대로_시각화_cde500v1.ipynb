```python
import json
import pandas as pd
import plotly.express as px
import os

PATH = "/content/cde_500v1.json"
```

```python
def load_section_scores_json(path: str) -> pd.DataFrame:
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    if isinstance(data, dict) and all(isinstance(v, dict) for v in data.values()):
        df = pd.DataFrame.from_dict(data, orient="index").reset_index().rename(columns={"index": "section"})
        return df

    if isinstance(data, list) and all(isinstance(v, dict) for v in data):
        df = pd.DataFrame(data)
        # 섹션 컬럼명이 다를 수 있음: "섹션", "title" 등을 "section"으로 통일
        for cand in ["section", "섹션", "title", "제목"]:
            if cand in df.columns:
                if cand != "section":
                    df = df.rename(columns={cand: "section"})
                break
        return df

    if isinstance(data, dict):
        # 안쪽에서 dict-of-dict 를 찾기
        for k, v in data.items():
            if isinstance(v, dict) and all(isinstance(x, dict) for x in v.values()):
                df = pd.DataFrame.from_dict(v, orient="index").reset_index().rename(columns={"index": "section"})
                return df

    raise ValueError("알 수 없는 JSON 구조입니다. 파일 내용을 확인하세요.")

# 1) 로드
PATH = "/content/cde_500v1.json"
df = load_section_scores_json(PATH)

# 2) 필요한 컬럼이 모두 있는지 확인/보정
expected_cols = ["coverage","accuracy","fluency","coherence","redundancy","avg_kpi","avg_len"]
for c in expected_cols:
    if c not in df.columns:
        df[c] = 0.0

# 3) long-form 변환
long_df = df.melt(
    id_vars="section",
    value_vars=expected_cols,
    var_name="eval",
    value_name="score"
)

# 4) 막대그래프 (그룹형)
fig = px.bar(
    long_df,
    x="section", y="score", color="eval",
    barmode="group",
    title="섹션별 평가 지표"
)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# 정규화 (redundancy는 낮을수록 좋으니 반전 후 정규화)
norm_cols = ["coverage","accuracy","fluency","coherence","redundancy","avg_kpi","avg_len"]
for col in norm_cols:
    if col == "redundancy":
        vals = -df[col]  # 반전
    else:
        vals = df[col]
    vmin, vmax = vals.min(), vals.max()
    if vmax > vmin:
        df[col+"_norm"] = (vals - vmin) / (vmax - vmin) * 100.0
    else:
        df[col+"_norm"] = 50.0  # 전부 같을 때 중간값

# 예: accuracy_norm 기준 상위/하위 5개
top5 = df.nlargest(5, "accuracy_norm")
bot5 = df.nsmallest(5, "accuracy_norm")

fig_top = px.bar(top5, x="section", y="accuracy_norm", title="Accuracy 상위 5개 (0–100 정규화)", text="accuracy_norm")
fig_top.update_traces(texttemplate='%{text:.1f}', textposition='outside')
fig_top.update_layout(xaxis_tickangle=-30, yaxis=dict(range=[0,100]))
fig_top.show()

fig_bot = px.bar(bot5, x="section", y="accuracy_norm", title="Accuracy 하위 5개 (0–100 정규화)", text="accuracy_norm")
fig_bot.update_traces(texttemplate='%{text:.1f}', textposition='outside')
fig_bot.update_layout(xaxis_tickangle=-30, yaxis=dict(range=[0,100]))
fig_bot.show()
```
