```python
# 머신러닝이 데이터를 학습하는 과정을 시각적으로 보여주는 예제

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 데이터 준비 
np.random.seed(42)
X = 2 * np.random.rand(50, 1)           # 독립변수 (예: 환자 나이)
y = 4 + 3 * X + np.random.randn(50, 1)  # 종속변수 (예: 혈압)
```

```python
# 모델  준비
model = LinearRegression()
```

```python
#  "학습 전" 상태 시각화
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="skyblue", label="trained data")
plt.title("Before: pre-trained")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.legend()
plt.show()
```

```python
# 모델 학습 (fit = 요리하기)
model.fit(X, y)
```

```python
# 예측 결과 
X_test = np.array([[0], [2]])
y_pred = model.predict(X_test)
```

```python
#  시각화
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="lightgray", label="trained data")
plt.plot(X_test, y_pred, color="tomato", linewidth=2, label="result")
plt.title("After: Model trained the pattern 모델이 데이터의 패턴을 학습한 모습")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.legend()
plt.show()
```

```python
print("모델이 학습한 레시피:")
print(f"절편(Intercept): {model.intercept_[0]:.2f}")
print(f"기울기(Slope): {model.coef_[0][0]:.2f}")
```
